#### High Correlation

```python
# Calculate correlation matrix
correlation_matrix = data.corr()

# Set a correlation threshold
correlation_threshold = 0.85

# Identify highly correlated pairs
highly_correlated_pairs = np.where(np.abs(correlation_matrix) > correlation_threshold)

# Create a set to keep track of features to remove
features_to_remove = set()

# Loop through the pairs and decide which features to remove
for feature1, feature2 in zip(*highly_correlated_pairs):
    if feature1 != feature2 and feature1 not in features_to_remove:
        features_to_remove.add(feature2)

# Convert indices to column labels
columns_to_remove = [data.columns[idx] for idx in features_to_remove]
```

```python
for column in data.columns[0:18]:
    plt.figure(figsize=(8,4))
    sns.boxplot(data=data, x=column)
    plt.show()
```

```python
# The interquartile range (IQR) is calculated to count how many outliers there are and possibly filter/remove them.

q1=np.percentile(data['air_time1'], 25)   # percentile 25
q3=np.percentile(data['air_time1'], 75)   # percentile 75
iqr=q3-q1 

upper_limit = q3 + 1.5*iqr
lower_limit = q1 - 1.5*iqr

average_balance_outliers = data[(data['air_time1']<lower_limit) | (data['air_time1']>upper_limit)]
print(len(average_balance_outliers))
print(data.air_time1.shape)
len(average_balance_outliers)/data.air_time1.shape[0]
```

```python
# Create a function to know how many outliers are, wich is the column shape 
# and wich is the pecentage that represents the outliers for every each column.

def number_of_outliers(df):
    outliers_data = []  # Empty list to store the data
    for column in df.columns:
        q1 = np.percentile(df[column], 25)  # Percentile 25
        q3 = np.percentile(df[column], 75)  # Percentile 75
        iqr = q3 - q1

        upper_limit = q3 + 1.5 * iqr
        lower_limit = q1 - 1.5 * iqr

        # Filter the outliers
        average_balance_outliers = df[(df[column] < lower_limit) | (df[column] > upper_limit)]
        num_outliers = len(average_balance_outliers)
        column_shape = df[column].shape
        percentage = (num_outliers / column_shape[0])*100

        # Append the data to the empty list
        outliers_data.append([column, num_outliers, percentage])

    # Create a DataFrame with the outliers data
    outliers_df = pd.DataFrame(outliers_data, columns=['column_name', 'number_outliers', '%'])
    
    return outliers_df
```

```python
outliers_df = number_of_outliers(data)
outliers_df.head()
```

### 4. Logistic Regression Model

```python
#Train test split

y = data['class']
X = data.drop(['class'], axis=1)
print(y.shape, X.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#Create and fit our regression model

lr = LogisticRegression(random_state=42, solver='lbfgs', max_iter=1000)
lr.fit(X_train, y_train)
```

```python
#Make predictions for train
y_train_pred = lr.predict(X_train)
y_train_pred
```

```python
#Make predictions for test
y_pred_test = lr.predict(X_test)
y_pred_test
```

##### -Evaluation of the model with the confusion matrix

```python
cm_train = confusion_matrix(y_train, y_train_pred)
cm_train
```

```python
#Display confusion matrix for train data
disp = ConfusionMatrixDisplay(cm_train, display_labels=lr.classes_)
disp.plot()
plt.show()
```

```python
#Confusion matrix using seaborn and percentages for train data
ax = sns.heatmap(cm_train/np.sum(cm_train), annot=True, fmt='.2%', cmap='Blues')
ax.set(xlabel="Predicted label", ylabel="True label")
plt.show()
```

```python
#Now for test
cm_test = confusion_matrix(y_test, y_pred_test)
disp = ConfusionMatrixDisplay(cm_test, display_labels=lr.classes_)
disp.plot()
plt.show()
```

```python
#Confusion matrix using seaborn and percentages for train data
ax = sns.heatmap(cm_test/np.sum(cm_test), annot=True, fmt='.2%', cmap='Blues')
ax.set(xlabel="Predicted label", ylabel="True label")
plt.show()
```

```python
print("The accuracy of the model in the TRAIN set is: {:.2f}".format(lr.score(X_train, y_train)))
print("The accuracy of the model in the TEST set is: {:.2f}".format(lr.score(X_test, y_test)))
```

When calculating only the accuracy score of the model, it indicates that it is very good with **89%** of the correct predictions, 
that is, the true positive and negative values. 

However, due to the type of error we want to minimize, we need to evaluate other metrics, such as `recall`, `precision`, `F1 score` and `cohen_kappa_score`

The type of error are:  
**Type I**: False Positives (FP): we <u>incorrectly</u> predict that people have Alzheimer's disease.  
**Type II**: False Negatives (FN): we <u>incorrectly</u> predict that people are healthy.  

In this case, we want to minimize Type II error because predicting that people are healthy when they are not,
can have serious consequences. 

Therefore, we need to improve <u>recall</u>.

```python
print("The accuracy in the TEST set is: {:.2f}".format(accuracy_score(y_test,y_pred_test)))
print("The precision in the TEST set is: {:.2f}".format(precision_score(y_test,y_pred_test)))
print("The recall in the TEST set is: {:.2f}".format(recall_score(y_test,y_pred_test)))
print("The F1 in the TEST set is: {:.2f}".format(f1_score(y_test,y_pred_test)))
print("The Kappa in the TEST set is: {:.2f}".format(cohen_kappa_score(y_test,y_pred_test)))
```

### 5. Improving the model

##### 5.1 High correlation columns removal (hr)

```python
data.shape
```

```python
# Create a new DataFrame without highly correlated features
hr_df = data.drop(columns=columns_to_remove)
hr_df.shape
```

```python
# Save data set without highly correlated columns 
hr_df.to_csv('../data/clean_data/data_without_highly_correlated_columns.csv', index=False)
```

```python
# Create a function to test multiple times the logistic regression model, trying to improve the metrics with different modifications

def logistic_regression_model(df):
    y = df['class']
    X = df.drop(['class'], axis=1)
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
    
    lr = LogisticRegression(random_state=0, solver='lbfgs', max_iter=1000)
    lr.fit(X_train, y_train)
    
    y_train_pred = lr.predict(X_train)
    y_pred_test = lr.predict(X_test)
    
    cm_test = confusion_matrix(y_test,y_pred_test)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm_test, display_labels=lr.classes_)
    disp.plot()
    
    accuracy = accuracy_score(y_test, y_pred_test)
    precision = precision_score(y_test, y_pred_test)
    recall = recall_score(y_test, y_pred_test)
    f1 = f1_score(y_test, y_pred_test)
    kappa = cohen_kappa_score(y_test, y_pred_test)
    
    result = f'''
    The accuracy in the TEST set is: {accuracy:.2f}
    The precision in the TEST set is: {precision:.2f}
    The recall in the TEST set is: {recall:.2f}
    The F1 in the TEST set is: {f1:.2f}
    The Kappa in the TEST set is: {kappa:.2f}
    '''

    for line in result.split('\n'):
        print(line)
```

```python
logistic_regression_model(hr_df)
```

The values of accuracy, recall, F1, and the Kappa <u>decreases</u>  
It **does not improve** the model by removing columns with high multicollinearity.

##### 5.4 Outliers imputation (oi)

```python
# Create a copy for the original dataframe to avoid make changes in it
oi_df = data
```

```python
outliers_index = {}  # A dictionary to store the indices of outliers for each column
modes = oi_df.mode().iloc[0]  # Calculate the mode for each column

# Using the mode to impute outliers may be more optimal, as it is less sensitive to these values, compared to using the mean.

# A for loop to find the interquantile range, the upper and lower limits, the indices for every each column, and impute outlier with the mode.
for column in oi_df.columns:
    q1 = np.percentile(oi_df[column], 25)  # Percentile 25
    q3 = np.percentile(oi_df[column], 75)  # Percentile 75
    iqr = q3 - q1

    upper_limit = q3 + 1.5 * iqr
    lower_limit = q1 - 1.5 * iqr

    # Find and store the indices of outliers
    outliers_index[column] = oi_df[(oi_df[column] < lower_limit) | (oi_df[column] > upper_limit)].index
    
    # Impute outliers with the mode
    oi_df.loc[outliers_index[column], column] = modes[column]
```

```python
# Save data set with the imputation of outliers 
oi_df.to_csv('../data/clean_data/data_with_the_imputation_of_outliers.csv', index=False)
```

```python
# Check the outliers with the boxplots.
for column in oi_df.columns[0:18]:
    plt.figure(figsize=(8,4))
    sns.boxplot(data=oi_df, x=column)
    plt.show()
```

```python
logistic_regression_model(oi_df)
```

After all outliers are imputed, <u>**all metrics decreases.**</u>

### 6. Conclusions

After conducting an `EDA` and implementing different actions to enhance the performance of the <u>*Logistic Regression Model*</u>, with the aim of reduce false negatives in predictions, we have identified the following points to highlight:

- <u>Decreasing model performance by removing highly correlated columns</u>. These columns can contain valuable information. Even though they are   highly correlated, they may contribute unique nuances to the model.  

- <u>The model's performance decreases by imputing outliers with the mode</u>. Outliers can contain important information about unusual or extreme data behavior. Imputing them can affect the model's ability to predict real-world scenarios.

- <u>When removing columns with low variance, the model's performance remains unchanged</u>. However, it reduces the complexity of the model and shortens the training time. 

- Finally, <u>by standardizing the data, the model's performance improves</u>. Logistic regression is sensitive to the scale of variables, and scaling the data ensures that all variables have a similar impact on the model, which can enhance its performance. 

The best metrics obtained in this model are the following: 

The `accuracy` in the TEST set is: **0.89**  
The `precision` in the TEST set is: **0.91**  
The `recall` in the TEST set is: **0.87**  
The `F1` in the TEST set is: **0.89**  
The `Kappa` in the TEST set is: **0.77**
